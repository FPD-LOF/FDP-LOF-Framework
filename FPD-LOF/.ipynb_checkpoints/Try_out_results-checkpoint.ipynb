# Plotting to compare outliers

```python
def compare_outlier(outlier,intersection,direction,results=result):
    """Function to plot the average FPD for an hour of a weekday against an outlier FPD"""
    outlier = pd.Timestamp(outlier)
    df = results[direction]['fpds'][intersection]
    outlier_data = df.loc[(df['timestamp'].dt.date ==outlier.date()) & ((df['timestamp'].dt.hour >=outlier.hour-1) & (df['timestamp'].dt.hour <=outlier.hour+1) )] #set apart the values for the hour of the outlier
    #now manipulate df to only contain FPDs for that weekday and hour; e.g. all flows for 18:00 at Thursdays.
    outlier_hours_df = pd.DataFrame()
    for i in range(-1,2): #also plot hour before and after
        a = df.loc[(df['timestamp'].dt.dayofweek ==outlier.dayofweek) & (df['timestamp'].dt.hour ==outlier.hour+i)] 
        a= a['prob'].groupby(by=a['timestamp'].dt.time).agg([np.mean, np.std])
        outlier_hours_df = pd.concat([outlier_hours_df,a],axis=0)
    
    xticks = [x for x in range(0,176,5)] #better index; values indicate the minutes (5,10, etc..)
    plt.plot(xticks,outlier_hours_df['mean'],label='Average FPD') #plot the average FPDs

    plt.fill_between(xticks, outlier_hours_df['mean']-outlier_hours_df['std'],outlier_hours_df['mean']+outlier_hours_df['std'], facecolor='#F0F8FF', alpha=1.0, edgecolor='none')
    plt.plot(xticks,outlier_data['prob'],label='Outlier FPD') #Plot the outlier FPD
    #plot hour seperators
    plt.axvline(60, color='black')
    plt.axvline(120, color='black')
    
    plt.legend()
    plt.title("Comparison of outlier on weekday: "+str(outlier.weekday())+" for time "+str(outlier))
    #plt.savefig('./figures/flow_compare_'+fpd_intersection+"_"+str(outlier.date())+"_"+str(outlier.hour)+".png")
    plt.show()
    return df, outlier_data, outlier_hours_df
```

```python
a,b, c = compare_outlier('2019-01-24 14:00:00','201234','NW')
```

```python
a,b, c = compare_outlier('2019-01-25 09:00:00','201234','NW')
```

```python
def compare_flows(flow1, flow2,intersection,direction,results=result):
    """Function to plot the average FPD for an hour of a weekday against an outlier FPD"""
    outlier = pd.Timestamp(flow1)
    df = results[direction]['fpds'][intersection]
    outlier_data = df.loc[(df['timestamp'].dt.date ==outlier.date()) & ((df['timestamp'].dt.hour==outlier.hour)  )] #set apart the values for the hour of the outlier
    #now manipulate df to only contain FPDs for that weekday and hour; e.g. all flows for 18:00 at Thursdays.
    outlier = pd.Timestamp(flow2)
    
    flow2_data = df.loc[(df['timestamp'].dt.date ==outlier.date()) & ((df['timestamp'].dt.hour==outlier.hour) )]
    outlier_hours_df = pd.DataFrame()
    for i in [0]: #also plot hour before and after
        a = df.loc[(df['timestamp'].dt.dayofweek ==outlier.dayofweek) & (df['timestamp'].dt.hour ==outlier.hour+i)] 
        a= a['prob'].groupby(by=a['timestamp'].dt.time).agg([np.mean, np.std])
        outlier_hours_df = pd.concat([outlier_hours_df,a],axis=0)
    
    xticks = [x for x in range(0,60,5)] #better index; values indicate the minutes (5,10, etc..)
    plt.plot(xticks,outlier_hours_df['mean'],label='Average FPD') #plot the average FPDs
    
    plt.fill_between(xticks, outlier_hours_df['mean']-outlier_hours_df['std'],outlier_hours_df['mean']+outlier_hours_df['std'], facecolor='#F0F8FF', alpha=1.0, edgecolor='none')
    plt.plot(xticks,outlier_data['prob'],label=pd.Timestamp(flow1)) #Plot the outlier FPD
    plt.plot(xticks,flow2_data['prob'],label=pd.Timestamp(flow1)) #Plot the outlier FPD
    print(outlier_data['prob'].values)
    print(flow2_data['prob'].values)
    #plot hour seperators
    #plt.axvline(60, color='black')
    #plt.axvline(120, color='black')
    plt.yticks(np.arange(0,0.13,0.02))
    plt.legend()
    plt.title("Comparison of flows at 10 AM on 05 and 19 November 2019")
    #plt.savefig('./figures/flow_compare_'+fpd_intersection+"_"+str(outlier.date())+"_"+str(outlier.hour)+".png")
    plt.show()
    return df, outlier_data, outlier_hours_df
```

```python
compare_flows("2018-11-05 10:00:00","2018-11-19 10:00:00","201234","NW")
```
