Implementation to create a json file for the tlc playbackservice. Should be an array nested with dicts per intersection. Each intersection has a name, virtualhost (NoordHolland) and then an array named 'Detectors', which has dicts with filename,name & start of data. 

```python
import os
```

```python
output = []
starttime="2018-09-14T00:00:00"
#intersections to include:
intersections=['201234','201231','201239','205195','201225','20149','201291','201308','205195']
for intersection in intersections:
    if intersection=="201234": #if control intersection add arrivals & pedestrian sensors.
        intersection_dict={"Name":intersection, "VirtualHost":"NoordHolland","Detectors":[]}
        if os.path.exists("./output/{}/arrival_detectors".format(intersection)):
            for file in os.listdir("./output/{}/arrival_detectors".format(intersection)):
                if file.endswith('.gz'):
                    intersection_dict['Detectors'].append({'FileName':intersection+"/"+file,
                                                          "Name":file.split(".")[1],
                                                          "StartDateOfData":starttime})

        if os.path.exists("./output/{}/pedestrians_cyclists".format(intersection)):
            for file in os.listdir("./output/{}/pedestrians_cyclists".format(intersection)):
                if file.endswith('.gz'):
                    intersection_dict['Detectors'].append({'FileName':intersection+"/"+file,
                                                          "Name":file.split(".")[1],
                                                          "StartDateOfData":starttime})
        output.append(intersection_dict)
    else: #for all other intersections add arrivals and stopline
        intersection_dict={"Name":intersection, "VirtualHost":"NoordHolland","Detectors":[]}
        if os.path.exists("./output/{}/arrival_detectors".format(intersection)):
            for file in os.listdir("./output/{}/arrival_detectors".format(intersection)):
                if file.endswith('.gz'):
                    intersection_dict['Detectors'].append({'FileName':intersection+"/"+file,
                                                          "Name":file.split(".")[1],
                                                          "StartDateOfData":starttime})

        if os.path.exists("./output/{}/stopline_detectors".format(intersection)):
            for file in os.listdir("./output/{}/stopline_detectors".format(intersection)):
                if file.endswith('.gz'):

                    intersection_dict['Detectors'].append({'FileName':intersection+"/"+file,
                                                          "Name":file.split(".")[1],
                                                          "StartDateOfData":starttime})
        output.append(intersection_dict)
```

```python
output
```

```python
import json
with open('./output/sept18_sept19_replay.json','w', encoding='utf-8') as out:
          json.dump(output,out,ensure_ascii=False, indent=4)
```
