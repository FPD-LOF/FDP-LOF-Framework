```python
import json
import pandas as pd
import matplotlib.pyplot as plt
#director= json.load(open("SG_info_CCOL_VER_TESTING_20190115_12_1hours(31).txt", 'rb'))
ccol=json.load(open("CCOL.txt", 'rb'))
t035=json.load(open("T035.txt", 'rb'))
org=json.load(open("original.txt", 'rb'))
#org_new=json.load(open("SG_info_CCOL_VER_TESTING_20190115_12_1hours(52).txt", 'rb'))
approaches= {'328':['1','2'],'334':['3','4'],'339':['5','6']}
```

```python
df = pd.DataFrame(columns=['CCOL_Wachten','T035_Wachten','Org_Wachten',
                          'CCOL_Stops','T035_Stops','Org_Stops'])
for mainkey in approaches.keys():
    for lane in approaches[mainkey]:
        df = df.append({'CCOL_Wachten':sum(ccol[mainkey][lane]['onnodige_wachttijd_SG']),                     
                        'T035_Wachten':sum(t035[mainkey][lane]['onnodige_wachttijd_SG']),
                        'Org_Wachten':sum(org[mainkey][lane]['onnodige_wachttijd_SG']),
                        'CCOL_Stops':sum(ccol[mainkey][lane]['stopcounter']),
                        'T035_Stops':sum(t035[mainkey][lane]['stopcounter']),
                        'Org_Stops':sum(org[mainkey][lane]['stopcounter']),}, ignore_index=True)
```

```python
waiting = pd.DataFrame()
stops = pd.DataFrame()
models = {"CCOL":ccol,"T035":t035,"Original":org}
for mainkey in approaches.keys():
    for lane in approaches[mainkey]:
        for model in models.keys():
            waiting[model+"_"+mainkey+"_"+lane+"_Wait"] = models[model][mainkey][lane]['onnodige_wachttijd_SG']
            stops[model+"_"+mainkey+"_"+lane+"_Stops"] = models[model][mainkey][lane]['stopcounter']

waiting['CCOL_avg'] = waiting[['CCOL_328_1_Wait','CCOL_328_2_Wait','CCOL_334_3_Wait','CCOL_334_4_Wait','CCOL_339_5_Wait','CCOL_339_6_Wait']].sum(axis=1)
waiting['T035_avg'] = waiting[['T035_328_1_Wait','T035_328_2_Wait','T035_334_3_Wait','T035_334_4_Wait','T035_339_5_Wait','T035_339_6_Wait']].sum(axis=1)
waiting['Original_avg'] = waiting[['Original_328_1_Wait','Original_328_2_Wait','Original_334_3_Wait','Original_334_4_Wait','Original_339_5_Wait','Original_339_6_Wait']].sum(axis=1)

stops['CCOL_avg'] = stops[['CCOL_328_1_Stops','CCOL_328_2_Stops','CCOL_334_3_Stops','CCOL_334_4_Stops','CCOL_339_5_Stops','CCOL_339_6_Stops']].sum(axis=1)
stops['T035_avg'] = stops[['T035_328_1_Stops','T035_328_2_Stops','T035_334_3_Stops','T035_334_4_Stops','T035_339_5_Stops','T035_339_6_Stops']].sum(axis=1)
stops['Original_avg'] = stops[['Original_328_1_Stops','Original_328_2_Stops','Original_334_3_Stops','Original_334_4_Stops','Original_339_5_Stops','Original_339_6_Stops']].sum(axis=1)
            
```

```python
stops[['CCOL_avg','T035_avg','Original_avg']].plot()
plt.title('Stops per minute')
```

```python
waiting[['CCOL_avg','T035_avg','Original_avg']].plot()
plt.title('Waiting time per minute')
```
